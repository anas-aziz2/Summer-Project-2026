```python
width_1 = 3
width_2 = 3
learn_rate = 0.01


W1 = np.random.rand(1, width_1)*np.sqrt(2/1)
W2 = np.random.rand(width_1, width_2)*np.sqrt(2/width_1)
W3 = np.random.rand(width_2, 1)*np.sqrt(2/width_2)

b1 = np.zeros(width_1)
b2 = np.zeros(width_2)
b3 = np.zeros(1)
```

```python
X=np.linspace(-np.pi, np.pi, 100).reshape(100,1)
Y=np.sin(X)
```

```python
Z2 =A1 @ W2 +b2
A2 = np.maximum(Z2, 0)
```

```python
Z3 =A2 @ W3 +b3
Y_pred = Z3
delta_3 = 2*(Y_pred - Y)/100
```
