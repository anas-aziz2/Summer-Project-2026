```python
import numpy as np
import torch
```

```python
tensor_A = torch.randn(3, 3)
print(tensor_A)
```

```python
tensor_B = torch.linalg.inv(tensor_A)
print(tensor_B)
```

```python
tensor_C = tensor_A @ tensor_B
print(tensor_C)
```

```python
tensor_A_grad=torch.gradient(tensor_A)
print("Gradient along rows:\n", tensor_A_grad[0])
print("Gradient along columns:\n", tensor_A_grad[1])
```
