```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
layer_widths = [1, 50, 50, 1]
learning_rate = 0.001
weights = []
biases = []
for i in range(len(layer_widths)-1):
    W = np.random.randn(layer_widths[i], layer_widths[i+1])*np.sqrt(2/layer_widths[i])
    b = np.zeros(layer_widths[i+1])
    weights.append(W)
    biases.append(b)
```

```python
X=np.linspace(-np.pi, np.pi, 100).reshape(100,1)
Y=np.sin(X)

```

```python
for epoch in range(500000):
    A=X
    activations = [A]
    Zs = []
    for W, b in zip(weights, biases):
        Z = A @ W + b
        A = np.maximum(Z, 0)
        Zs.append(Z)
        activations.append(A)

    Y_pred = Zs[-1]
    deltas = []
    for i in reversed(range(len(weights))):
        if i == len(weights) - 1:
            delta = 2 * (Y_pred - Y) / X.shape[0]
        else:
            delta = deltas[-1] @ weights[i + 1].T * (Zs[i] > 0)
        
        dW = (activations[i].T @ delta)
        db = np.sum(delta, axis=0)
        weights[i] -= learning_rate * dW
        biases[i] -= learning_rate * db
        deltas.append(delta)


plt.scatter(X, Y, label='True')
plt.scatter(X, Zs[-1], label='Predicted')
```

```python
# Effect of Learning Rate:
# For a given netwwork, as the learning rate increases, the model fits faster but at some point, the weights become too large for the calculation to proceed. 
# Another observation is that, low learning rates can still lead to model fitting in the expense of learning time (no. of epochs)

#Effect of Network Width and Depth
# As the width and depth increases, the network fits better, however, increase in depth requires more time for the calcualtions. Also, increasing depth by 20 
# neurons shows superior fitting when compared to increasing the depth by 20 neurons.  
```
