```python
import numpy as np
```

```python
# f(x,y) = x^2 + xy + y^2
x = 2
y = 3
lr = 0.01

x_values = [2]
y_values = [3]
grad_x_values = []
grad_y_values = []


for i in range(5):
    grad_x = 2*x + y
    grad_y = x + 2*y

    x = x - lr * grad_x
    y = y - lr * grad_y

    x_values.append(x)
    y_values.append(y)
    grad_x_values.append(grad_x)
    grad_y_values.append(grad_y)

print(grad_x_values)
print(grad_y_values)
```

```python
import torch
```

```python
x_auto = torch.tensor(2.0,requires_grad=True)
y_auto = torch.tensor(3.0,requires_grad=True)
grad_x_auto = []
grad_y_auto = []

for i in range(5):
    f = x_auto**2 + x_auto*y_auto + y_auto**2

    f.backward()

    with torch.no_grad():
        x_auto -= lr*x_auto.grad
        y_auto -= lr*y_auto.grad

    grad_x_auto.append(x_auto.grad.item())
    grad_y_auto.append(y_auto.grad.item())
    x_auto.grad.zero_()
    y_auto.grad.zero_()

print(grad_x_auto)
print(grad_y_auto)
```

```python
# The gradients do match! 
# f.backward() seems to be a powerful tool to automatically calculate the gradients. This will make the backprop step a lot easier
```
