Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Define Material Properties

```python
k = 237
cp = 900
rho = 2700

alpha = k / (cp * rho)

```

Geometry and Discretization

```python
L = 2
div = 100
del_x = L / div

t = 100000
del_t = 1

```

Boundary Conditions

```python
T0 = 400
TL = 20

T=np.ones((t+1, div+1)) * TL
T[:,0]=T0
T[:,-1]=TL
```

Governing Equation Solving

```python
for i in range(1,t+1):
    for j in range(1,div):
        T[i,j] = T[i-1,j] + alpha * del_t / del_x**2 * (T[i-1,j+1] - 2*T[i-1,j] + T[i-1,j-1])

```

Plotting

```python
fig, axs = plt.subplots(3,sharex=True)
plt.subplots_adjust(hspace=0.9)
fig.suptitle('TRANSIENT HEAT CONDUCTION IN A ROD')
x = np.linspace(0, L, div+1)
axs[0].plot(x,T[0,:])
axs[1].plot(x, T[1000,:])
axs[2].plot(x, T[100000,:])
axs[0].set_title('t=0s')
axs[1].set_title('t=1000s')
axs[2].set_title('t=100000s')
axs[2].set_xlabel('Position along the length of the rod (m)')
fig.supylabel('Temperature (C)')
```
